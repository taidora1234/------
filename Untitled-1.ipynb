```python
import pandas as pd
import requests

keyword = "Андеррайтер"
russia = 113
per_page = 100
pages = 19

df_list = []

for page in range(1, pages + 1):
    url = f'https://api.hh.ru/vacancies?text={keyword}&area={russia}&per_page={per_page}&page={page}'
    #url = f'https://api.hh.ru/vacancies?text={keyword}&per_page={per_page}&page={page}'
    response = requests.get(url)
    data = response.json()
    vacancies = data['items']

    for vacancy in vacancies:
        published_at = pd.to_datetime(vacancy['published_at'])
        name = vacancy['name']
        employer = vacancy['employer']['name']
        area = vacancy['area']['name']
        salary = vacancy['salary']

        if salary is None:
            salary_from = None
        else:
            salary_from = salary.get('from')
            
        responsibility = vacancy['snippet']['responsibility']
        requirement = vacancy['snippet']['requirement']
        employment = vacancy['employment']['name']
        experience = vacancy['experience']['name']

        if responsibility is not None:
            responsibility = responsibility.replace("<highlighttext>", "")
            responsibility = responsibility.replace("</highlighttext>", " ")
        if requirement is not None:
            requirement = requirement.replace("<highlighttext>", "")
            requirement = requirement.replace("</highlighttext>", " ")            

        df_list.append({
            'published_at': published_at,
            'name': name,
            'employer': employer,
            'area': area,
            'salary': salary,
            'responsibility': responsibility,
            'requirement': requirement,
            'employment': employment,
            'experience': experience
        })

df = pd.DataFrame(df_list)
```

```python
import pandas as pd
from collections import Counter
from nltk.tokenize import word_tokenize

# Combine all requirement strings into a single text
all_requirements = ' '.join(df['requirement'].dropna())

# Tokenize the text into words and phrases
words_and_phrases = word_tokenize(all_requirements)

# Remove irrelevant phrases and words
irrelevant_phrases = ['не менее', 'х лет', 'работы', 'в области', '1 года']
filtered_phrases = [phrase for phrase in words_and_phrases if phrase not in irrelevant_phrases]

# Count the occurrences of each phrase
phrase_counts = Counter(filtered_phrases)

# Select the top 5 most frequent phrases
top_phrases = phrase_counts.most_common(5)

# Print the top 5 most frequent phrases
for phrase, count in top_phrases:
    print(phrase, count)
```
