```python
import pandas as pd
import requests

keyword = "Андеррайтер"
#moscow = 1
per_page = 100
pages = 10

df_list = []

for page in range(1, pages + 1):
    #url = f'https://api.hh.ru/vacancies?text={keyword}&area={moscow}&per_page={per_page}&page={page}'
    url = f'https://api.hh.ru/vacancies?text={keyword}&per_page={per_page}&page={page}'
    response = requests.get(url)
    data = response.json()
    vacancies = data['items']

    for vacancy in vacancies:
        published_at = pd.to_datetime(vacancy['published_at'])
        name = vacancy['name']
        employer = vacancy['employer']['name']
        salary = vacancy['salary']

        if salary is None:
            salary_from = None
        else:
            salary_from = salary.get('from')
            
        responsibility = vacancy['snippet']['responsibility']
        requirement = vacancy['snippet']['requirement']
        employment = vacancy['employment']['name']
        experience = vacancy['experience']['name']

        if responsibility is not None:
            responsibility = responsibility.replace("<highlighttext>", "")
            responsibility = responsibility.replace("</highlighttext>", " ")
        if requirement is not None:
            requirement = requirement.replace("<highlighttext>", "")
            requirement = requirement.replace("</highlighttext>", " ")            

        df_list.append({
            'published_at': published_at,
            'name': name,
            'employer': employer,
            'salary': salary,
            'responsibility': responsibility,
            'requirement': requirement,
            'employment': employment,
            'experience': experience
        })

df = pd.DataFrame(df_list)
```

```python
df['salary'] = df['salary'].apply(lambda x: x['from'] if isinstance(x, dict) else None)
df['published_at'] = pd.to_datetime(df['published_at']).dt.date
```

```python
import pandas as pd
from collections import Counter
from nltk import ngrams
import re

# Function to get phrases from text
def get_phrases(text, n):
    if text is None:
        return []
    words = re.findall(r'\b\w+\b', text.lower())
    return list(ngrams(words, n))

# Applying the function to the "requirement" column to get 3-word phrases
n = 3
phrases = df['requirement'].apply(lambda x: get_phrases(x, n)).explode().tolist()

# Counting the frequency of phrases
phrase_counts = Counter(phrases)

# Getting the most common phrases
top_phrases = phrase_counts.most_common(10)

# Printing the most common phrases
for phrase, count in top_phrases:
    print(f'Phrase: {" ".join(phrase)}, Count: {count}')

```
